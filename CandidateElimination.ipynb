```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("https://github.com/rajathv/Machine-Learning-Lab/raw/master/program2/trainingexamples.csv")
data.head()
```

```python
x = data.values.tolist()
```

```python
def candidateE(x):
  specific = x[0].copy()
  specific.pop()
  general = [['?' for i in range(len(specific))] for i in range(len(specific))]

  for h in x:
    if h[-1] =="Y":
      for i in range(len(specific)):
        if h[i]!=specific[i]:
          specific[i]='?'
          general[i][i]='?'

    if h[-1] == "N":
      for i in range(len(specific)):
        if h[i]!=specific[i]:
          general[i][i] = specific[i]
        else:
          general[i][i] = '?'

  count=0
  for v in general:
    if v==['?', '?', '?', '?', '?', '?']:
      count+=1
  for i in range(count):
    general.remove(['?', '?', '?', '?', '?', '?'])

  return specific, general
```

```python
s_final, g_final = candidateE(x)
print("Final Specific_h:", s_final, sep="\n")
print("Final General_h:", g_final, sep="\n")
```
