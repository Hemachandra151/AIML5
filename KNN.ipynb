#Train Test Split

```python
x,tx=train_test_split(data,random_state=2529)
key=['Iris-setosa','Iris-versicolor','Iris-virginica']
farm=x.values.tolist()
```

#K-NN Classifier

```python
def knn(sl,sw,pl,pw):
  k=15
  n=len(farm)
  d=np.full(k,np.nan)
  norm=np.zeros(n)
  freq=np.zeros(3)

  for i in range(n):
    sld=farm[i][0]-sl
    swd=farm[i][1]-sw
    pld=farm[i][2]-pl
    pwd=farm[i][3]-pw

    norm[i]=math.sqrt(sld*sld+swd*swd+pld*pld+pwd*pwd)

  count=0
  while count<k:
    mini=9999
    for i in range(n):
      if norm[i]<mini and i not in d:
        mini=norm[i]
        val=i
    d[count]=val
    count+=1

  for i in range(k):
    freq[key.index(farm[int(d[i])][4])]+=1

  return key[np.argmax(freq)]
```

#Testing The Model

```python
N=len(tx)
sofa=tx.values.tolist()
conf = np.zeros((3,3))
for i in range(N):
  prd=knn(sofa[i][0],sofa[i][1],sofa[i][2],sofa[i][3])
  if prd=='Iris-setosa':
    if sofa[i][4]=='Iris-setosa':
      conf[0][0]+=1
    if sofa[i][4]=='Iris-versicolor':
      conf[0][1]+=1
    if sofa[i][4]=='Iris-virginica':
      conf[0][2]+=1
  elif prd=='Iris-versicolor':
    if sofa[i][4]=='Iris-setosa':
      conf[1][0]+=1
    if sofa[i][4]=='Iris-versicolor':
      conf[1][1]+=1
    if sofa[i][4]=='Iris-virginica':
      conf[1][2]+=1
  elif prd=='Iris-virginica':
    if sofa[i][4]=='Iris-setosa':
      conf[2][0]+=1
    if sofa[i][4]=='Iris-versicolor':
      conf[2][1]+=1
    if sofa[i][4]=='Iris-virginica':
      conf[2][2]+=1

print(f"Out of {conf[0][0]+conf[1][0]+conf[2][0]} classified as Iris-setosa {conf[0][0]} were correctly predicted")
print(f"Out of {conf[0][1]+conf[1][1]+conf[2][1]} classified as Iris-versicolor {conf[1][1]} were correctly predicted")
print(f"Out of {conf[0][2]+conf[1][2]+conf[2][2]} classified as Iris-virginica {conf[2][2]} were correctly predicted")
n = conf[0][0]+conf[1][1]+conf[2][2]
print(f"{n} were correctly classified out of {N}")
print(f"Accuracy:{n*100/N}")
```
