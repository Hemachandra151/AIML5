#Train Test Split

```python
x,tx = train_test_split(data, random_state=2529)
target=x[['medv']].values.tolist()
```

#Data Preprocessing

```python
col=['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat']
lx=x[col].values.tolist()
ltx=tx[col].values.tolist()
for v in lx:
  v.append(1)
for v in ltx:
  v.append(1)
lx = np.array(lx)
ltx = np.array(ltx)
```

#Locally Weighted Regression

```python
n= len(lx)
W = np.zeros((n,n))
t=11

def lwr(p):
  for i,v in enumerate(lx):
    norm = np.dot((v-p),(v-p))
    norm/=(2*t)
    W[i][i] = np.exp(-norm)

  L = np.transpose(lx) @ W @ lx + 1e-5 * np.identity(14)
  R = np.transpose(lx) @ W @ target
  P = np.linalg.inv(L) @ R
  P = P.reshape(-1)

  return np.dot(P,p)
```

#Testing The Model

```python
N = len(ltx)
out = tx[['medv']].values.tolist()
prd = np.zeros(N)
act = np.zeros(N)
rms = 0

for i,v in enumerate(out):
  act[i] = v[0]

for i,v in enumerate(ltx):
  prd[i] = lwr(v)
  rms+=(prd[i]-act[i])*(prd[i]-act[i])

rms = np.sqrt(rms/N)
print(f"Mean Square Error {rms}")
```

#Visualization of Actual vs Predicted

```python
import matplotlib.pyplot as plt
plt.scatter(act,prd)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```
